```r
dataset=read.table("/sharedFolder/Results/Idelalisib/HEATMAPFigure/count_hgtgts.csv",header=TRUE,row.names=1,sep=",")
```

```r
library(preprocessCore)

# Assicurati che dataset sia un data.frame numerico
data_matrix <- as.matrix(dataset)

# Applica Quantile Normalization
normalized_matrix <- normalize.quantiles(data_matrix)

# Ripristina nomi originali
colnames(normalized_matrix) <- colnames(dataset)
rownames(normalized_matrix) <- rownames(dataset)

# Converti a dataframe per semplicitÃ 
normalized_dataset <- as.data.frame(normalized_matrix)

# Scrivi il file CSV
write.csv(normalized_dataset, file = "/sharedFolder/Data/dataset_quantile_normalized.csv", row.names = TRUE)

# Conferma
print("Quantile normalization eseguita e salvata correttamente.")
```

```r
quantile_normalize <- function(df){
    df_rank <- apply(df, 2, rank, ties.method = "min")  # rank delle colonne
    sorted_df <- apply(df, 2, sort)                     # ordina ogni colonna
    mean_sorted <- rowMeans(sorted_df)                  # media per ogni riga (quantile medio)

    # sostituisci i valori originali con i valori quantile-normalizzati
    df_normalized <- apply(df_rank, 2, function(ranks) mean_sorted[ranks])

    # ripristina i nomi originali
    colnames(df_normalized) <- colnames(df)
    rownames(df_normalized) <- rownames(df)

    return(as.data.frame(df_normalized))
}
```

```r
# Uso della funzione sul dataset
dataset_normalized <- quantile_normalize(dataset)

# Salva il risultato in CSV
write.csv(dataset_normalized, file = "/sharedFolder/Results/Idelalisib/HEATMAPFigure/dataset_quantile_normalized_manual.csv", row.names = TRUE)
```
