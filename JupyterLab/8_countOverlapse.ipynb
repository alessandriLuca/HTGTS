```r
library("GenomicRanges")
 library(gtools)
path="/sharedFolder/Data/1_HTGTS/8_BCELL_LAM_AID_KO//TEST2//pooledDiff/"
```

```r
myc=GRanges(seqnames="chr1",ranges=IRanges(start=192576266,end=192581297))
a=mixedsort(list.files(path))
overlaps=c()
for(i in a){
e=read.table(paste(path,"/",i,sep=""),header=FALSE,sep="\t")
e=GRanges(seqnames=e[,1],ranges=IRanges(start=e[,2],end=e[,3]))
overlaps=append(overlaps,(countOverlaps(myc,e)))
}
names(overlaps)=a
overlaps
```
