```r
library("GenomicRanges")
 library(gtools)
myc=GRanges(seqnames="chr15",ranges=IRanges(start=60562682,end=63076130))
a=mixedsort(list.files("/sharedFolder/Data/1_HTGTS/1_Junction_Jianli/junction_singles/"))
junctionNumber=c()
percentageBS=c()
for(i in a){
e=read.table(paste("/sharedFolder/Data/1_HTGTS/1_Junction_Jianli/junction_singles/",i,sep=""),header=FALSE,sep="\t")
db=GRanges(seqnames=as.matrix(e[,1]),ranges=IRanges(start = as.numeric(e[,2])-100,end=as.numeric(e[,3])+100))
percentageBS=append(percentageBS,length(findOverlaps(myc,db)@to)/nrow(e))
junctionNumber=append(junctionNumber,nrow(e))    
}
```

```r
noExt=function(x){
return(tools::file_path_sans_ext(x))
}
fNumber=function(x){
sapply(x,FUN=function(x){
strsplit(x,"__")[[1]][2]
})
}
```

```r
djMat=cbind(percentageBS,junctionNumber)
rownames(djMat)=noExt(a)
```

```r
summa=c()
for(i in unique(fNumber(noExt(a)))){
tempSamples=length(grep(i,rownames(djMat)))
tempSum=sum(djMat[grep(i,rownames(djMat)),2])
summa=append(summa,rep(tempSum,tempSamples))    
}
```

```r
length(summa)
```

```r
djMat=cbind(djMat,summa)
colnames(djMat)=c("% of junctions around breaksite(±2Mb)","Junction number","Total junction number in replicates")
```

```r
No=as.numeric(sapply(rownames(djMat),FUN=function(x){strsplit(x,"__")[[1]][1]}))
```

```r
CellType=sapply(rownames(djMat),FUN=function(x){strsplit(x,"__")[[1]][1]})
```

```r
cellType=sapply(rownames(djMat),FUN=function(x){strsplit(strsplit(x,"__")[[1]][2],"_")[[1]][1]})
```

```r
Treatment=sapply(rownames(djMat),FUN=function(x){strsplit(strsplit(x,"__")[[1]][2],"_")[[1]][2]})
```

```r
djMat2=cbind(No,noExt(a),cellType,Treatment,"c-myc 25xI-SceI",junctionNumber,percentageBS,summa)
```

```r
colnames(djMat2)=c("No.","Sample","Cell Type","Treatment","Breaksite","Junction number","% of junctions around breaksite(±2Mb)","Total junction number in replicates")
```
