```r
dmso_poso <- read.table(paste("/sharedFolder/Data/5_CutTagAce/MACS2_220607_JT10123_MN00260_trimmed_","DMSO","/macs2_peak_q0.1_peaks.csv",sep=""), header = TRUE, sep = ",")
taze_poso <- read.table(paste("/sharedFolder/Data/5_CutTagAce/MACS2_220607_JT10123_MN00260_trimmed_","Taze","/macs2_peak_q0.1_peaks.csv",sep=""), header = TRUE, sep = ",")
vale_poso <- read.table(paste("/sharedFolder/Data/5_CutTagAce/MACS2_220607_JT10123_MN00260_trimmed_","Vale","/macs2_peak_q0.1_peaks.csv",sep=""), header = TRUE, sep = ",")
```

```r
# Carica la libreria
library(GenomicRanges)

# Crea oggetti GRanges per ciascun DataFrame e aggiungi i valori di abs_summit
dmso_gr <- with(dmso_poso, GRanges(seqnames = as.character(chr), ranges = IRanges(start-5000, end+5000), summit = abs_summit))
taze_gr <- with(taze_poso, GRanges(seqnames = as.character(chr), ranges = IRanges(start-5000, end+5000), summit = abs_summit))
vale_gr <- with(vale_poso, GRanges(seqnames = as.character(chr), ranges = IRanges(start-5000, end+5000), summit = abs_summit))

# Unisci gli oggetti GRanges
merged_gr <- reduce(c(dmso_gr, taze_gr, vale_gr))


```

```r
length(merged_gr)
```

```r
# Estrai i valori di summit dai GRanges originali
summit_dmso <- as.numeric(mcols(dmso_gr)$summit)
summit_taze <- as.numeric(mcols(taze_gr)$summit)
summit_vale <- as.numeric(mcols(vale_gr)$summit)

# Inizializza vettori vuoti per i valori di summit di ogni campione
summit_dmso_merged <- numeric(length(merged_gr))
summit_taze_merged <- numeric(length(merged_gr))
summit_vale_merged <- numeric(length(merged_gr))

# Itera attraverso ogni coordinata in merged_gr
for (i in seq_along(merged_gr)) {
  coordinata <- merged_gr[i]
  
  # Verifica se la coordinata è presente in dmso_gr, taze_gr, o vale_gr
  if (coordinata %in% dmso_gr) {
    summit_dmso_merged[i] <- summit_dmso[which(coordinata %in% dmso_gr)]
  }
  if (coordinata %in% taze_gr) {
    summit_taze_merged[i] <- summit_taze[which(coordinata %in% taze_gr)]
  }
  if (coordinata %in% vale_gr) {
    summit_vale_merged[i] <- summit_vale[which(coordinata %in% vale_gr)]
  }
}
```

```r
totMat=cbind(summit_dmso_merged,summit_taze_merged,summit_vale_merged)
write.table(totMat,"/sharedFolder/Data/5_CutTagAce/peaks_totMat.csv",col.names=NA)
```
