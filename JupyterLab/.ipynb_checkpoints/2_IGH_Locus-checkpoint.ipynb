```r
# centering with 'scale()'
center_scale <- function(x) {
    scale(x, scale = FALSE)
}
library("pheatmap")
results="/sharedFolder/Results/Hotspot"
dir.create(results)
previousResults="/sharedFolder/Results/Hotspot/"
library("ggVennDiagram");
library("ggplot2")
library("vioplot")
path="/sharedFolder/Data/1_HTGTS/3_JunctionPaper/junction_pooled"
library(GenomicRanges)
```

```r
type=c("WT","Idelalisib","AID_KO","Ligase4")
conditions=c("DMSO","Taze","Vale")
ff=list.files(path)

igh1=GRanges(seqnames="chr12",ranges=IRanges(start = 114499552-5500,end=114503186+5500))
igh2=GRanges(seqnames="chr12",ranges=IRanges(start = 114550646-5500,end=114552106+5500))
igh3=GRanges(seqnames="chr12",ranges=IRanges(start = 114573701-5500,end=114576919+5500))
igh4=GRanges(seqnames="chr12",ranges=IRanges(start = 114601912-5500,end=114603131+5500))
igh5=GRanges(seqnames="chr12",ranges=IRanges(start = 114662040-5500,end=114664834+5500))

mm=matrix(nrow=5)
names=c()
for(i in type){
   ff2=ff[grep(i,ff)]   
for(j in conditions){
    ff3=ff2[grep(j,ff2)]
names=append(names,paste(i,j,sep="_"))
junctions=read.table(paste(path,ff3,sep="/"),header=FALSE,sep="\t")
db=GRanges(seqnames=as.matrix(junctions[,1]),ranges=IRanges(start = as.numeric(junctions[,2]),end=as.numeric(junctions[,3])))
mm=cbind(mm,c(countOverlaps(igh1,db),countOverlaps(igh2,db),countOverlaps(igh3,db),countOverlaps(igh4,db),countOverlaps(igh5,db)))

}
}
mm=mm[,-1]
colnames(mm)=names
rownames(mm)=c("igh1","igh2","igh3","igh4","igh5")
write.table(mm,paste(results,"/igh_junctions.csv",sep=""),col.names=NA,sep=",")
```
