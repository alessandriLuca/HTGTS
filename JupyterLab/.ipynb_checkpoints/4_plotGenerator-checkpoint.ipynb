```r
# centering with 'scale()'
center_scale <- function(x) {
    scale(x, scale = FALSE)
}
library("pheatmap")
results="/sharedFolder/Results/All_hotspot/Data_Analysis/WT"
results2="/sharedFolder/Results/All_hotspot/Data_Analysis/Lig4"
previousResults="/sharedFolder/Results/Hotspot/"
```

```r
tempHeat=read.table(paste(results,"/Taze.csv",sep=""),header=TRUE,sep=",",row.names=1)
Database=unique(read.table(paste(previousResults,"Database.csv",sep="/"),header=TRUE,sep=",")[,c(1,2,3,7)])
rownames(Database)=Database[,4]
DBTU=Database[rownames(tempHeat),]
bigOne=read.table(paste(results,"/rpm_hgtgtsHeatmap.csv",sep=""),header=TRUE,sep=",",row.names=1)
bigTwo=read.table(paste(results,"/groseq_Heatmap.csv",sep=""),header=TRUE,sep=",",row.names=1)
bigThree=read.table(paste(results,"/countRAW_hgtgtsHeatmap.csv",sep=""),header=TRUE,sep=",",row.names=1)
bigFour=read.table(paste(results2,"/countRAW_hgtgtsHeatmap.csv",sep=""),header=TRUE,sep=",",row.names=1)
rmarkdown=matrix(ncol=1)
rmarkdown=rbind(rmarkdown,"---")
rmarkdown=rbind(rmarkdown,'title: "Hotspots"')
rmarkdown=rbind(rmarkdown,"---")
#write batch file 
batch=matrix(ncol=1)
countings=matrix(ncol=2)
countings2=matrix(ncol=2)
countings3=matrix(ncol=2)
countings4=matrix(ncol=2)

colnames(countings)=c("HotspotBelonging","RPM")
colnames(countings2)=c("HotspotBelonging","gro-seq")
colnames(countings3)=c("HotspotBelonging","countRaw")
colnames(countings4)=c("HotspotBelonging","countRaw")

for(i in seq(1,nrow(DBTU))){
temp=(DBTU[i,3]-DBTU[i,2])*10
batch=rbind(batch,paste("goto ",DBTU[i,1],":",(DBTU[i,2]-temp),"-",(DBTU[i,3]+temp),sep=""))
batch=rbind(batch,paste("snapshot ",i,".png",sep=""))
t=cbind(colnames(bigOne),t(bigOne[rownames(DBTU)[i],]))
t2=cbind(colnames(bigTwo),t(bigTwo[rownames(DBTU)[i],]))
t3=cbind(colnames(bigThree),t(bigThree[rownames(DBTU)[i],]))
t4=cbind(colnames(bigFour),t(bigFour[rownames(DBTU)[i],]))

colnames(t)=c("HotspotBelonging","RPM")
rownames(t)<-NULL
countings=rbind(countings,t)

colnames(t2)=c("HotspotBelonging","gro-seq")
rownames(t2)<-NULL
countings2=rbind(countings2,t2)
    
colnames(t3)=c("HotspotBelonging","countRaw")
rownames(t3)<-NULL
countings3=rbind(countings3,t3)
    
colnames(t4)=c("HotspotBelonging","countRaw")
rownames(t4)<-NULL
countings4=rbind(countings4,t4)
#markdown
    
johnTemp=" "
for(sd in seq(1,nrow(t))){
johnTemp=paste(johnTemp,"<br>",paste(t[sd,],collapse=" "),sep="")
}
for(sd in seq(1,nrow(t2))){
johnTemp=paste(johnTemp,"<br>",paste(t2[sd,],collapse=" "),sep="")
}
for(sd in seq(1,nrow(t3))){
johnTemp=paste(johnTemp,"<br>",paste(t3[sd,],collapse=" "),sep="")
}
    
for(sd in seq(1,nrow(t4))){
johnTemp=paste(johnTemp,"<br>",paste(t4[sd,],collapse=" "),sep="")
}
rmarkdown=rbind(rmarkdown,paste("## ",rownames(DBTU)[i],sep=""))
rmarkdown=rbind(rmarkdown,paste("![",johnTemp,"](./",i,".png",")",sep=""))
rmarkdown=rbind(rmarkdown," ")

}
batch=batch[-1,]
write.table(batch,paste(results,"/","Taze.batch",sep=""),col.names=FALSE,row.names=FALSE,sep="\t",quote=FALSE)
rmarkdown=rmarkdown[-1,]
write.table(rmarkdown,paste(results,"/","Taze.rmd",sep=""),row.names=FALSE,col.names=FALSE,quote=FALSE)
```
