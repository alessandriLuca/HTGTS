```r
#Path
expName="HATCR"
samples=c("M234")
#outputs
outPath="/sharedFolder/jupyterlab_outputs/"
threshold=4
prefixPath="/sharedFolder/scratch/results/"
prefixPath2="/outs/per_sample_outs/"
suffixRNA="/count/sample_filtered_feature_bc_matrix/"
```

```r
#Libraries
library(Matrix)
library(data.table)
library(DESeq2)
library("edgeR")
library(tibble)
library(umap)
library("AnnotationDbi")
library(tsne)
library("org.Mm.eg.db")
library("clusterProfiler")

```

```r
#clustering.Output_creation
for(i in samples){
#i=samples[1]
    outPath2=paste0(outPath,"/",i)
outPath3=paste0(outPath,"/",i,"/RNA_seq")
outPath4=paste0(outPath,"/",i,"/RNA_seq/ExpansVSExpans")
dir.create(outPath4)
barcodes=read.table(paste(outPath2,"/",i,"_ByNucleotidesCLM.csv",sep=""),sep=",",header=TRUE)
#            ADAPTEEEEEEEEER
tmpbar=c()
for(clo in unique(barcodes[,1])){
tmpbar=append(tmpbar,which(barcodes[,1]==clo)[1])
}
barcodes=barcodes[tmpbar,]
rownames(barcodes)=barcodes[,1]
#              ADAPTEEEEEEEEER
interesting=rownames(barcodes)[which(barcodes[,2]>threshold)]
blist=list()
count=1
clustering.output=matrix(ncol=2)
for(j in interesting){
blist[[count]]=barcodes[j,22:ncol(barcodes)][barcodes[j,22:ncol(barcodes)]!=""]
clustering.output=rbind(clustering.output,cbind(blist[[count]],j))
count=count+1
}
clustering.output=clustering.output[-1,]
colnames(clustering.output)=c("blist","cl")
write.table(clustering.output,paste(outPath4,"/",i,"_clustering.output_Expans.csv",sep=""),row.names=FALSE,sep=",")   

}
```

```r
set.seed(123)
for(i in samples){
outPath2=paste0(outPath,"/",i)
outPath3=paste0(outPath,"/",i,"/RNA_seq")
outPath4=paste0(outPath,"/",i,"/RNA_seq/ExpansVSExpans")
outPath5=paste0(outPath,"/",i,"/RNA_seq/ExpansVSExpans/temp/")
dir.create(outPath5)
geneList=c("PDCD1","LAG3","TNF","IFNG","GZMB","IL2")
dir.create(outPath4)
countMatrix=read.table(paste(outPath3,"/",i,"_countMatrix.csv",sep=""),header=TRUE,row.names=1,sep=",",check.names = FALSE)
clustering.output=read.table(paste(outPath4,"/",i,"_clustering.output_Expans.csv",sep=""),header=TRUE,sep=",")
scoreMatAverage=matrix(ncol=length(geneList))    
    for(j in unique(clustering.output[,2])){
        scoreMat=matrix(ncol=length(geneList))
        names=c()
        for(k in unique(clustering.output[,2])){
            if(j!=k){
                tmpCL=clustering.output[clustering.output[,2] %in% c(j),]
                tmpCL[,2]="B"
                tmpCL2=clustering.output[clustering.output[,2] %in% c(k),]
                tmpCL2[,2]="A"
                tmpCL=rbind(tmpCL,tmpCL2)
                tmpcountMatrix=countMatrix[,tmpCL[,1]]
                dge <- DGEList(counts=tmpcountMatrix, group=as.factor(tmpCL[,2]))
                dge_norm <- calcNormFactors(dge)
                design <- model.matrix(~group, data=dge_norm$samples)
                dge_norm <- estimateDisp(dge_norm, design, robust=TRUE)
                fit <- glmFit(dge_norm, design)
                lrt <- glmLRT(fit)
                topTags(lrt)
                ttt=lrt$table[sort(lrt$table$logFC,index.return=TRUE,decreasing=TRUE)$ix,]
                ttt=ttt[which(abs(ttt$PValue)<0.05),]
                mapped_genes=mapIds(org.Mm.eg.db, keys=rownames(ttt), keytype="ENSEMBL", column="SYMBOL")
                tempGene=c()
                for(gene in geneList){
                    PDCD1=ttt[which(toupper(mapped_genes)==gene),]$logFC
                    if(length(PDCD1)==0){
                        PDCD1=0
                        }
                    tempGene=append(tempGene,PDCD1)
                    
                }
                scoreMat=rbind(scoreMat,tempGene) 
                names=append(names,paste(j,"VS",k,sep=""))
            }
        }
        scoreMat=scoreMat[-1,]
        rownames(scoreMat)=names
        colnames(scoreMat)=geneList
        write.table(scoreMat,paste(outPath5,"/",j,".csv",sep=""),col.names=NA,sep=",")
        scoreMatAverage=rbind(scoreMatAverage,colMeans(scoreMat))
    #MEDIA DI SCOREMAT per J   
    }
scoreMatAverage=scoreMatAverage[-1,]
rownames(scoreMatAverage)=unique(clustering.output[,2])
write.table(scoreMatAverage,paste(outPath4,"/",i,"_scoreAverage.csv",sep=""),col.names=NA,sep=",")
  }  
```
