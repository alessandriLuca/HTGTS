```r
#SETTING PATH 
path="/sharedFolder/Data/HTGTS/4_Junction_Jianli_final/junction_pooled/"
pathGroseq="/sharedFolder/Data/gro-seq/"
previousResults="/sharedFolder/Results/Hotspot/"
results="/sharedFolder/Results/All_hotspot/Data_Analysis"
dir.create(results,recursive=TRUE)
riderPath="/home/Riderv2/"
```

```r
#LOAD USEFULL FUNCTIONS 
noExt=function(x){
return(tools::file_path_sans_ext(x))
}
bnoExt=function(x){
return(basename(tools::file_path_sans_ext(x)))
}
fNumber=function(x){
sapply(x,FUN=function(x){
strsplit(x,"__")[[1]][2]
})
}
rRider=function(x){
sapply(x,FUN=function(x){
strsplit(x,"-RIDER")[[1]][1]
})
}
# centering with 'scale()'
center_scale <- function(x) {
    scale(x, scale = FALSE)
}
```

```r
#LOAD LIBRARIES 
library("argparser");
library(gtools);
library("GenomicRanges");
library("ggVennDiagram");
options(warn=-1)
library(GenomicRanges)
```

```r
#Load Data
Database=read.table(paste(previousResults,"Database.csv",sep="/"),header=TRUE,sep=",")
a=list.files(path,pattern=".bed",recursive=TRUE)
a=mixedsort(a)
a2=sapply(a,FUN=function(x){strsplit(x,".bed")[[1]][1]})
a2=sapply(a2,FUN=function(x){strsplit(x,"__")[[1]][2]})
```

```r
print("WT ANALYSIS")
results="/sharedFolder/Results/All_hotspot/Data_Analysis/WT"
dir.create(results,recursive=TRUE)
```

```r
Database=Database[-grep("Idelalisib",Database[,6]),]
Database=Database[grep("WT",Database[,6]),]
a2=a2[-grep("Idelalisib",a2)]
a2=a2[grep("WT",a2)]
a=a[-grep("Idelalisib",a)]
a=a[grep("WT",a)]
```

```r
#Heatmap matrix generation for HTGTS data
hgtgts_count=matrix(0,ncol=length(unique(Database[,7])),nrow=length(a2))
colnames(hgtgts_count)=sapply(unique(Database[,7]),FUN=toString)
rownames(hgtgts_count)=a2
hgtgts_rpm=hgtgts_count
hgtgts_fpkm=hgtgts_count
for(j in a){
    system(paste("echo",j))
temp=read.table(paste(path,"/",j,sep=""),sep="\t")
tempGG=GRanges(seqnames=as.matrix(temp[,1]),ranges=IRanges(start = as.numeric(temp[,2]),end=as.numeric(temp[,3])))    
    for(i in seq(nrow(Database))){
        system(paste("echo",i/nrow(Database)))
        gg=GRanges(seqnames=as.matrix(Database[i,1]),ranges=IRanges(start = as.numeric(Database[i,2]),end=as.numeric(Database[i,3])))    
        over=countOverlaps(gg,tempGG)
        name=as.vector(noExt(fNumber(j)))
        hgtgts_count[name,toString(Database[i,7])]= over
        hgtgts_rpm[name,toString(Database[i,7])]= ((over)/nrow(temp))*1000000
        hgtgts_fpkm[name,toString(Database[i,7])]= ((((over)/nrow(temp))*1000000)/(Database[i,3]-Database[i,2]))*1000000
   }

}
#write.table(log2(t(hgtgts_count)+1),paste(results,"/","count_hgtgtsHeatmap.csv",sep=""),col.names=NA,sep=",")
#write.table(log2(t(hgtgts_rpm)+1),paste(results,"/","rpm_hgtgtsHeatmap.csv",sep=""),col.names=NA,sep=",")
#write.table(log2(t(hgtgts_fpkm)+1),paste(results,"/","fpkm_hgtgtsHeatmap.csv",sep=""),col.names=NA,sep=",")


#write.table((t(hgtgts_count)),paste(results,"/","countRAW_hgtgtsHeatmap.csv",sep=""),col.names=NA,sep=",")

#hgtgts_count=log2(t(hgtgts_count)+1)
#hgtgts_rpm=log2(t(hgtgts_rpm)+1)
#hgtgts_fpkm=log2(t(hgtgts_fpkm)+1)
```

```r

#Heatmap matrix generation for gro-seq data
Database=read.table(paste(previousResults,"Database.csv",sep="/"),header=TRUE,sep=",")
Database=Database[-grep("Idelalisib",Database[,6]),]
Database=Database[grep("WT",Database[,6]),]
a=list.files(pathGroseq,pattern=".bed",recursive=TRUE)
a=mixedsort(a)
a2=sapply(a,FUN=function(x){strsplit(x,".bed")[[1]][1]})
groseq_count=matrix(0,ncol=length(unique(Database[,7])),nrow=length(a2))
colnames(groseq_count)=sapply(unique(Database[,7]),FUN=toString)
rownames(groseq_count)=a2
for(j in a){
    system(paste("echo",j))
temp=read.table(paste(pathGroseq,"/",j,sep=""),sep="\t")
tempGG=GRanges(seqnames=as.matrix(temp[,1]),ranges=IRanges(start = as.numeric(temp[,2]),end=as.numeric(temp[,3])),mcols=abs(as.numeric(temp[,4])))
    for(i in seq(nrow(Database))){
        system(paste("echo",i/nrow(Database)))
        gg=GRanges(seqnames=as.matrix(Database[i,1]),ranges=IRanges(start = as.numeric(Database[i,2]),end=as.numeric(Database[i,3])))    
        over=findOverlaps(gg,tempGG)@to    
        name=as.vector(noExt((j)))
        groseq_count[name,toString(Database[i,7])]= sum(tempGG[over]$mcols)
   }

}
write.table((t(groseq_count)),paste(results,"/groseq_Heatmap.csv",sep=""),col.names=NA,sep=",")
```
