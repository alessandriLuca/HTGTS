```r
# ðŸ“Œ VARIABLES

# A variable is like a box that stores a value.
# In R, you can assign values using = 

name = "Alice"
surname = "Rossi"
age = 23
grade = 27.5

# Let's print them
print(name)
print(age)
print(grade)
```

```r
# ðŸ“Œ FUNCTIONS

# A function is a block of code that does something specific.
# Think of it like a small machine: you give it input, it gives you output.
# Very important: "What happens in a function stays in the function"
# (Like Las Vegas ðŸ˜„ â€“ variables created inside the function donâ€™t exist outside)

# Let's define a function that adds two numbers
add_numbers = function(a, b) {
  result = a + b
  return(result)
}

# Now we call the function
sum_result = add_numbers(5, 7)

# Print the result
print(sum_result)

# Let's try to print 'result' outside the function
# This will give an error because 'result' only exists inside the function
# print(result)  # = Uncommenting this line will give an error
```

```r
# ðŸ“Œ WORKING WITH STUDENTS' DATA

# Let's say we want to store information for 3 students:
# - First name
# - Last name
# - Grade

# METHOD 1: Using separate variables for each student (not practical for many students)

name1 = "Alice"
surname1 = "Rossi"
grade1 = 28

name2 = "Marco"
surname2 = "Bianchi"
grade2 = 30

name3 = "Sara"
surname3 =  "Verdi"
grade3 = 25

print(paste(name1, surname1, "-", grade1))
print(paste(name2, surname2, "-", grade2))
print(paste(name3, surname3, "-", grade3))
```

```r
# METHOD 2: Using vectors (more efficient)

names = c("Alice", "Marco", "Sara")
surnames = c("Rossi", "Bianchi", "Verdi")
grades = c(28, 30, 25)

# Let's access the first student's name
print(names[1])  # Alice
print(surnames[1])  # Rossi
print(grades[1])  # 28
```

```r
# METHOD 3: Using a for loop to print all students' information

for (i in 1:length(names)) {
  full_name = paste(names[i], surnames[i])
  grade = grades[i]
  print(paste(full_name, "- Grade:", grade))
}
```

```r
# A global variable (defined outside any function)
x <- 10

# Define a function that adds 5 to a number
add_five <- function(y) {
  # 'y' is a parameter: it only exists inside the function
  z <- y + 5   # 'z' is a local variable: only visible inside the function
  return(z)
}

# Call the function with x as input
result <- add_five(x)

# Print the result
print(result)  # Output: 15

# Try to print 'z' outside the function
# print(z)  # This will cause an error: 'z' does not exist here

# Check that x has not changed
print(x)  # Still 10
#Key Concepts:
#Global variable: defined outside a function, visible everywhere (x).
#Function parameter: a variable that exists only inside the function (y).
#Local variable: created inside the function and disappears after the function ends (z).
#The function does not modify global variables unless you use special syntax (<<-, which is not recommended for beginners).
```

```r
bonus <- 2  # global variable

add_bonus <- function(score) {
  # The function can read the global variable 'bonus'
  final <- score + bonus
  return(final)
}

print(add_bonus(28))  # Output: 30
#Functions can access global variables, but they donâ€™t change them unless explicitly told to.
```

```r
# Create 3 vectors (like columns of a table)
names <- c("Alice", "Marco", "Sara")
surnames <- c("Rossi", "Bianchi", "Verdi")
grades <- c(28, 30, 25)

# Loop through each student
for (i in 1:length(names)) {
  # i is the index of the current student (1, 2, 3...)

  current_name <- names[i]
  current_surname <- surnames[i]
  current_grade <- grades[i]

  full_name <- paste(current_name, current_surname)

  print(paste(full_name, "- Grade:", current_grade))
}
#What Happens in Each Iteration?
#Letâ€™s break it down:

#First iteration (i = 1):
#names[1] â†’ "Alice"
#surnames[1] â†’ "Rossi"
#grades[1] â†’ 28
#â†’ Printed: Alice Rossi - Grade: 28
#Second iteration (i = 2):
#names[2] â†’ "Marco"
##surnames[2] â†’ "Bianchi"
#grades[2] â†’ 30
#â†’ Printed: Marco Bianchi - Grade: 30
#Third iteration (i = 3):
#names[3] â†’ "Sara"
#surnames[3] â†’ "Verdi"
#grades[3] â†’ 25
#â†’ Printed: Sara Verdi - Grade: 25

```
